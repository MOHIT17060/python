```python
a=int(input("enter number"))
even_total=0
odd_total=0
for i in range(1,(a*2)+1):
    if i%2==0:
        even_total=i+even_total
        
    else:
        odd_total=i+odd_total
        
print(even_total,"even")
print(odd_total,"odd")
        
```

```python
#wpp to check whwther the no is prime or not,take user input

n =int(input("enter number"))
if n ==0 or n==1:
    print(n,"not a prime")
elif n>1:
    for i in range(2,n):
        if n%i==0:
            print(n,"not a prime")
            break
    else:
        print(n,"prime")
```

```python
# wpp to print prime number between 10 and 20
s=10
e=20
for i in range(s,e+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i,end=" ") 
        
```
