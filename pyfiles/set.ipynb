```python
s=set()
print(s)
```

```python
s={1,2,3,"hello"}
print(s)
print(type(s))
```

```python
#add method
s={1,2,3,"hello"}
s.add(1000)
print(s)
```

```python
#direct iteration
s={1,2,3,"hello"}
print(len(s))
for i in s:
    print(i,end=" ")
```

```python
#pop method
s={1,2,3,"hello"}
p=s.pop()
print(s)
print(p)
```

```python
#remove method
s={1,2,3,"hello"}
p=s.remove(15)
print(s)
```

```python
#discard method
s={1,2,3,"hello"}
p=s.discard(15)
print(s)
```

```python
#clear method
s={1,2,3,"hello"}
s.clear()
print(s)
```

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}
print(s1.union(s2))
```

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}
print(s1.intersection(s2))
```

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}
print(s1.difference(s2))
```

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}
print(s1.symmetric_difference(s2))
```
