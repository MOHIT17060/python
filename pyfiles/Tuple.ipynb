```python
l=[1,5,3,4,5,1,6,7,7,5,9,11,5]
l1=sorted(l)
l2=int(len(l)//2)
if l2%2!=0:
    for i in range(l2-1,l2):
        l3=(l1[i]+l1[i+1])/2
    print(l3)
else:
    for i in range(l2,l2+1):
        print(l1[i])
```

```python
sorted_list=sorted(l)
n=len(sorted_list)
if n%2!=0:
    median=sorted_list[n//2]
else:
    median=(sorted_list[n//2-1]+sorted_list[n//2]/2)
print(median)             
```
