# operators

```python
#operators
#arithmetic operators

a=10
b=2
print("addition :",a+b)
print("subtraction :",a-b)
print("multiplication :",a*b)
print("division :",a/b)
print("floor division :",a//b)
print("modulus :",a%b)
print("exponential power :",a**b)
```

### assignment operators

```python
a=10
b=8
c=a+b
a+=b
a-=10
a*=5
a//=5
b*=10
print(a)
print(b)
print(c)
```

### comparison operators == , <= ,>= ,!= , < ,>

```python
a=10
b=8

print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
print(a>b)
print(a<b)





```

### logical operators  and/or

```python
a=10
b=8
c=12
print(a>b and b>c)
print(a>b or b>c)
print(a>b and not b>c)
```

### identity operators     is , is not , id() = return memory address of an object
  

```python
a =10
b =8
print(id(a))
print(id(b))
print(a is b)
print(a is not b )
```

### membership operator       in/not in

```python
a = "hello"
print("h" in a)
print("h" not in a)
l=(10,20,"hello")
print("h" in l)
```

### bitwise 

```python
# &  -  bitwise and 
# | - bitwise or 
# ^ - bitwise xor 
# bin

a=10
b=8
print(a,"",bin(a))
print(b,"",bin(b))
print(bin(a&b)) 
print(a&b)
print(bin(a|b))
print(a|b)
print(bin(a^b))
print(a^b)      
```
