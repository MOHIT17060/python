# STRING

```python
# IT IS AN ORDERED SEQUENCE OF CHARACTER 
# 1. SINGLE LINE - DENOTED BY ' ' OR "" ""
# 2. MULTI LINE - DENOTED BY ''' ''' 
# INDEXED AND IMMUTABLE
s="" #empty string
print(s)
print(type(s))
```

```python
                 # SINGLE LINE


#indexed
b="decode"
print(b[0])

#immutable
s="decode"
print(s)
print(type(s))
print(s[0])
s[0] = "z"
print(s)
```

```python
                     #MULTILINE
#indexed
s='''HELLO 
       WELCOME 
            TO 
              BHOPAL'''
print(s)
print(s[0])

#immutable
d='''HELLO 
       WELCOME 
            TO 
              BHOPAL'''
d[0] = "r"
print(d)
```

# LIST

```python
# IT IS A COLLECTION OF ELEMENT 
# IT IS DENOTED BY [] WITH CSV
# INDEXED, ORDEREWD , MUTUABLE AND HETEROGENEOUS IN Nature 
l=[] #empty
print(l)
print(type(l))
```

```python
l=[10,20,"decode",10.5]
print(l)
print(l[3]) #indexed
l[0]="bhopal"
print(l)
```

# TUPLE

```python
#COLLECTION OF ELEMENT 
# DENOTED BY () WITH CSV 
# ORDERED , INDEXED , IMMUTABLE AND HETEROGENEOUS
t =() #empty tuple
print(t)
print(type(t))
```

```python
# SINGLE VALUE TUPLE 
t=(10,)
print(t)
```

```python
# multi value tuple
t=(10,20,300,"decode",10.5)
print(t)
print(t[2])
t[2]=400 #immutuable
print(t)
```

# dictionary

```python
# IT IS A COLLECTION OF DATAVALUE 
# ORDERED , UNINDEXED AND MUTUABLE 
# KEYS ------ UNIQUE
# VALUES ------ DUPLICATE 
d={}
print(d)
print(type(d))



```

```python
d={"name":"abc","subject":"da","fees":"25000"}
print(d)
print(d["name"])
d["name"]="def"
print(d)
```

# set
