```python
#iteration
d={"name":["abc","def","xyz"],"courses":["da","ds","de"],"fees":[30000,25000,40000]}
for i in d:
    print(d[i])
```

```python
d={"name":["abc","def","xyz"],"courses":["da","ds","de"],"fees":[30000,25000,40000]}
for i in d:
    for j in d[i]:
        print(j)
```

```python
#get
#dictobj.get(key):
d={"name":["abc","def","xyz"],"courses":["da","ds","de"],"fees":[30000,25000,40000]}
print(d.get("name"))
for i in d.get("name"):
    print(i)
    
```

```python
#keys
#dictobj.keys()
d={"name":["abc","def","xyz"],"courses":["da","ds","de"],"fees":[30000,25000,40000]}
print(d.keys())
for i in d.keys():
    print(i)
```

```python
#values
#dict.obj.values()
d={"name":["abc","def","xyz"],"courses":["da","ds","de"],"fees":[30000,25000,40000]}
print(d.values())
for i in d.values():
    print(i)
```

```python
#items
#dictobj.items()
d={"name":["abc","def","xyz"],"courses":["da","ds","de"],"fees":[30000,25000,40000]}
print(d.items())
for i in d.items():
    print(i)
```

```python
d={"name":["abc","def","xyz"],"courses":["da","ds","de"],"fees":[30000,25000,40000]}
print(d.items())
for i,j in d.items():
    print(i,j)
```
